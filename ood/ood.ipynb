```python
from foundation_models.chat_openai import AIModelType
from foundation_models.llama_oodt import LlamaOOTD
from foundation_models.mistral_oodt import MistralOOTD
from foundation_models.claude_oodt import ClaudeOODT
from foundation_models.openai_oodt import ChatOpenAIOutOfDomainDetection

df_ood = pd.read_json("../../dataset/ood_queries.json").sample(frac=1)

model_name = AIModelType.CLAUDE_OPUS
chat_model = ClaudeOODT(model=model_name)
response = await chat_model.generate_response("Mein Tablet ist kaputt, ich will eine neues kaufen")
print(response)
```

```python
import pandas as pd
from sklearn.metrics import classification_report, accuracy_score
from foundation_models.chat_openai import AIModelType
from foundation_models.mistral_oodt import MistralOOTD


model_name = AIModelType.CLAUDE_SONNET
# Initialize the model
chat_model = ClaudeOODT(model=model_name)

# Load and shuffle the dataset
df_ood_sample = df_ood

# Create lists to store predictions and true labels
predicted_ood = []

# Iterate through the dataset and make predictions
for index, row in df_ood_sample.iterrows():
    query = row['query']
    true_ood = row.get('outOfDomain')
    
    # Get the model's prediction
    response = await chat_model.generate_response(query)
    print(response)
    
    # Store the predictions
    predicted_ood.append(response.get('outOfDomain'))

# Add the predictions to the dataframe for comparison
df_ood_sample['predicted_outOfDomain'] = predicted_ood

# Overall evaluation: In-Domain vs Out-of-Domain classification
print("=== Overall In-Domain vs Out-of-Domain Classification ===")
accuracy_ood = accuracy_score(df_ood_sample['outOfDomain'], df_ood_sample['predicted_outOfDomain'])
print(f"Overall Accuracy: {accuracy_ood:.2f}")
print(classification_report(df_ood_sample['outOfDomain'], df_ood_sample['predicted_outOfDomain'], target_names=["In-Domain", "Out-of-Domain"]))


# write the results to a file
df_ood_sample.to_json(f"../../dataset/ood_queries_{model_name}_with_predictions.json", orient="records")
```
